```python
import requests
import pandas as pd
```

Fetching data from API and contaning it in response

```python
response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=e03cc31660c9febef44f01350fb3c1e6').json()
```

Now using loop we can fetch data as per our needs

```python
for i in response['results']:
  print(i)
```

We will now create a data frame and display our data their

```python
#made empty list
popularity =[]
vote_count=[]
original_language=[]
title = []
vote_average= []
overview=[]
release_date=[]

for i in response['results']:
  popularity.append(i['popularity'])
  vote_count.append(i['vote_count'])
  original_language.append(i['original_language'])
  title.append(i['title'])
  vote_average.append(i['vote_average'])
  overview.append(i['overview'])
  release_date.append(i['release_date'])


  #Now we will make a dictionary
  d={'title': title,'original_language':original_language,'overview':overview,'release_date': release_date,'popularity': popularity,'vote_count':vote_count,'vote_average': vote_average}
df=pd.DataFrame(d)
df
```

Now we have to do this for 500 more data, where we have to logically append all of them below the previous

```python
#jo response page 1 pe tha usko j tak chalana hai toh page wala part modify kr diye using .formt function
final = pd.DataFrame()
all_dfs=[] #creating a list to store all dataframes
for j in range(1,501):
  response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=e03cc31660c9febef44f01350fb3c1e6&page={}'.format(j)).json()
  popularity =[]
  vote_count=[]
  original_language=[]
  title = []
  vote_average= []
  overview=[]
  release_date=[]

  for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_language.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])


    #Now we will make a dictionary
    d={'title': title,'original_language':original_language,'overview':overview,'release_date': release_date,'popularity': popularity,'vote_count':vote_count,'vote_average': vote_average}
  df=pd.DataFrame(d)
  all_dfs.append(df)

final = pd.concat(all_dfs, ignore_index=True)
```

Data cleaning krenge

```python
#avg vote count jo hai usko ek decimal tak krna hai
final['vote_average'] = final['vote_average'].apply(lambda x: round(x, 1))
final
```

```python
# Convert 'release_date' column to datetime format first
final['release_date'] = pd.to_datetime(final['release_date'], errors='coerce')

# Now, convert 'release_date' to 'dd-mm-yyyy' format
final['release_date'] = final['release_date'].dt.strftime('%d-%m-%Y')
final
```

```python
print(final.isnull().sum())
```

```python
# Forward fill missing 'release_date' values
final['release_date'] = final['release_date'].fillna(method='ffill')
```
